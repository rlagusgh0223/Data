```python
ns_df = pd.read_csv('ns_202404.csv', low_memory=False)
ns_df.head()
```

이럴 때는 **불리언 배열**을 사용하면 쉽게 원하는 열만 선택할 수 있다

```python
# 데이터프레임 열 이름이 저장된 columns 속성 확인
print(ns_df.columns)
```

```python
# columns 속성은 Index 클래스 객체이다
# 파이썬의 리스트처럼 쓸 수 있다
print(ns_df.columns[0])
print(ns_df.columns[1])
print(ns_df.columns[-1])
```

```python
# 원소별 비교 : 리스트의 모든 원소와 비교해준다
ns_df.columns != 'Unnamed: 13'
```

```python
# 중간의 부가기호 열 제외
selected_columns = ns_df.columns != '부가기호'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
# 열 삭제하기
# drop(삭제하려는 열 이름, axis=1) 메서드
# axis 는 삭제할 축을 지정할 수 있다
# 기본값 0은 행, 1은 열을 의미한다
ns_book = ns_df.drop('Unnamed: 0', axis=1) # 이거 두 줄로 하면 밑에 것만 돌아간다
ns_book = ns_df.drop('Unnamed: 13', axis=1)
ns_book.head()
```

```python
# 제외할 열 이름을 리스트 형식으로 여러개 지정
ns_book = ns_df.drop(['Unnamed: 0', '부가기호', 'Unnamed: 13'], axis=1)
ns_book.head()
```

```python
# inplace=True로 지정하면 선택한 데이터프레임에 바로 수정된다
ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
```

dropna() 메서드   
NaN이 하나 이상 포함된 행이나 열을 삭제한다

```python
ns_book = ns_df.dropna(axis=1)
ns_book.head()
```

```python
# how='all' : NaN인 열을 삭제
ns_book = ns_df.dropna(axis=1, how='all')
ns_book.head()
```

행 삭제하기  
axis 매개변수 기본값이 0이기 때문에 생략해도 된다

```python
ns_book2 = ns_book.drop([0, 1])
ns_book2.head()
```

```python
 # [] 연산자와 슬라이싱
 ns_book2 = ns_book[2:]
 ns_book2.head()
```

```python
ns_book2 = ns_book[0:2]
ns_book2.head()
```

```python
# []연산자와 불리언 배열
selected_rows = ns_df['출판사'] == '한빛미디어'# 출판사가 '한빛미디어'인 행만 True인 불리언 배열
ns_book2 = ns_book[selected_rows]
ns_book2.head()
```

```python
# 대출건수가 1000이하인 행 삭제
ns_book2 = ns_book[ns_book['대출건수'] > 1000]
ns_book2.head()
```

중복된 행 찾기

```python
# duplicated() 메서드를 쓰면 두번째 중복된 값 부터는 True가 된다
sum(ns_book.duplicated())  # 번호가 다 다르기 때문에 중복된 행이 없다
```

```python
# sunset 매개변수 : 일부 열을 기준으로 중복된 행 찾는다
sum(ns_book.duplicated(subset=['도서명','저자','ISBN']))
```

```python
# 중복된 데이터 확인
# keep=False : 중복된 행을 모두 True로 반환
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN'], keep=False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

그룹별로 모으기

```python
count_df = ns_book[['도서명','저자','ISBN', '권', '대출건수']]
group_df = count_df.groupby(by=['도서명','저자','ISBN', '권'], dropna=False)  # dropna=False : NaN이 있는 행을 삭제하지 않는다
loan_count = group_df.sum()  # 같은 책의 대출건수는 하나로 합친다
```

```python
# 위의 두 작업을 하나로 이어서 함
loan_count = count_df.groupby(by=['도서명','저자','ISBN', '권'], dropna=False).sum()
loan_count.head()
```

원본 데이터 업데이트하기

```python
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN', '권'])  # 중복된 행을 True로 표시
unique_rows = ~dup_rows  # 불리언 배열을 반전시켜 고유한 행을 True로 표시
ns_book3 = ns_book[unique_rows].copy()  # 고유한 행만 선택
```

```python
sum(ns_book3.duplicated(subset=['도서명','저자','ISBN', '권']))  # 중복된 행 확인
```

원본 데이터프레임 인덱스 설정

```python
# 열을 인덱스로 설정 : set_index()
ns_book3.set_index(['도서명','저자','ISBN', '권'], inplace=True)
ns_book3.head()
```

```python
# 업데이트하기
ns_book3.update(loan_count)
ns_book3.head()
```

```python
# 인덱스 재설정 : reset_index()
ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

```python
sum(ns_book['대출건수'] > 100)
```

```python
sum(ns_book4['대출건수'] > 100)
```

```python
# 열의 순서를 바꾸는 방법 : []연산자에 열 이름을 순서대로 전달한다
ns_book4 = ns_book4[ns_book.columns]  # ns_book 데이터프레임의 열 이름을 전달한다
ns_book4.head()
```

```python
ns_book4.to_csv('ns_book4.csv', index=False)
```

# 03-2 잘못된 데이터 수정하기

```python
import pandas as pd
ns_book4 = pd.read_csv('ns_book4.csv', low_memory=False)
ns_book4.head()
```

```python
ns_book4.info()
```

```python
# 비어 있는 값을 셀 때는 isna()메서드로 불리언 배열을 반환한 후
# sum() 메서드로 True 개수를 세면 된다
ns_book4.isna().sum()
```

```python
# isna() 메서드는 누락된 값을 확인하고
# notna() 메서드는 누락되지 않은 값을 확인한다
ns_book4.notna().sum()
```

```python
ns_book4.loc[0, '도서권수'] = None
ns_book4['도서권수'].isna().sum()
```

```python
ns_book4.head(2)
```

```python
ns_book4.loc[0, '도서권수'] = 1
# 원래 판다스는 NaN을 실수로 받는데 아래 입력을 통해 int(정수)로 출력하게 한다
ns_book4 = ns_book4.astype({'도서권수':'int32', '대출건수':'int32'})
ns_book4.head(2)
```

```python
ns_book4.loc[0, '부가기호'] = None
ns_book4.head(2)
```

```python
# 데이터프레임 셀 값을 NaN으로 입력하려면 np.nan을 이용한다
import numpy as np
ns_book4.loc[0, '부가기호'] = np.nan
ns_book4.head(2)
```

누락된 값 바꾸기(1) : loc, fillna() 메서드

```python
set_isbn_na_rows = ns_book4['세트 ISBN'].isna()  # 누락된 값을 찾아불리언 배열로 반환
ns_book4.loc[set_isbn_na_rows, '세트 ISBN'] = ''  # 누락된 값을 빈 문자열로 바꿈
ns_book4['세트 ISBN'].isna().sum()  # 누락된 값이 몇 개인지 계산
```

```python
ns_book4.fillna('없음').isna().sum()  # fillna()메서드에 원하는 값을 NaN에서 바꿈
```

```python
ns_book4['부가기호'].fillna('없음').isna().sum()  # 부가기호 열만 선택
```

```python
# NaN을 바꾸면서 전체 데이터프레임 반환
ns_book4.fillna({'부가기호':'없음'}).isna().sum()
```

누락된 값 바꾸기(2) : replace()메서드  
replace메서드는 NaN 뿐만 아니라 어떤 값도 바꿀 수 있다

```python
# 바꾸려는 값이 하나일 때
# replace(원래 값, 새로운 값)
ns_book4.replace(np.nan, '없음').isna().sum()
```

```python
# 바꾸려는 값이 여러 개일 때
# replace([원래 값1, 원래 값2], [새로운 값1, 새로운 값2])
ns_book4.replace([np.nan, '2024'], ['없음', '24']).head(2)
```

```python
# {원래 값1:새로운 값1, 원래 값3:새로운 값2} 처럼 딕셔너리 형식으로 전달할 수 도 있다
ns_book4.replace({np.nan:'없음', '2024':'24'}).head()
```

```python
# 열 마다 다른 값으로 바꿀 때
# replace({열 이름: 원래 값}, 새로운 값)
ns_book4.replace({'부가기호': np.nan}, '없음').head(2)
```

```python
# 중첩된 딕셔너리로 전달
ns_book4.replace({'부가기호': {np.nan: '없음'},
                  '발행년도': {'2024': '24'}}).head(2)
```

정규표현식

```python
# regex=True : 정규표현식 사용한다는 의미
ns_book4.replace({'발행년도': {r'\d\d(\d\d)': r'\1'}}, regex=True)[100:102]
# r은 정규 표현식을 다른 문자열과 구분하기 위한 접두사
```

```python
# 중괄호를 이용하여 사용 개수를 지정할 수 있다
ns_book4.replace({'발행년도': {r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]
```

```python
# 발행년도 열에서 숫자가 아닌 문자를 포함하는 모든 행 검색
invalid_number = ns_book4['발행년도'].str.contains('\D', na=True)
print(invalid_number.sum())  # 숫자 이외의 문자가 들어간 행의 개수 출력
ns_book4[invalid_number].head()
```

```python
# 연도 앞과 뒤에 있는 문자 제외
ns_book5 = ns_book4.replace({'발행년도':r'.*(\d{4}).*'}, r'\1', regex=True)
ns_book5[invalid_number].head()
```

```python
# 숫자 이외의 문자가 들어간 행의 개수와 데이터 확인(860 -> 91개로 줄었다)
unknown_year = ns_book5['발행년도'].str.contains('\D', na=True)
print(unknown_year.sum())
ns_book5[unknown_year].head()
```

```python
# 이제 변환되지 않은 값은 NaN이거나 네 자리 숫자가 아닌 값이다
# 임의로 -1로 바꾸고 astype() 메서드로 '발행년도' 열의 데이터 타입을 정수형인 int32로 반환한다
ns_book5.loc[unknown_year, '발행년도'] = '-1'
ns_book5 = ns_book5.astype({'발행년도': 'int32'})
```

```python
# 4000년이 넘는 연도에서 2333을 빼서 서기로 바꾼 다음 연도가 4000년이 넘는 도서가 있는지 확인
yy_rows = ns_book5['발행년도'].gt(4000)  # 4000보다 큰 값을 찾는다
ns_book5.loc[yy_rows, '발행년도'] = ns_book5.loc[yy_rows, '발행년도'] - 2333
yyear = ns_book5['발행년도'].gt(4000)  # 다시 4000보다 큰 값을 찾는다
print(yyear.sum())
ns_book5[yyear].head(2)
```

```python
# 연도가 높은 도서 14권은 -1로 표기한다
ns_book5.loc[yyear, '발행년도'] = -1
```

```python
# 연도가 0보다 크고 1900보다 작은 값
old_books = ns_book5['발행년도'].gt(0) & ns_book5['발행년도'].lt(1900)
ns_book5[old_books]
```

```python
# 잘못된 값은 -1로 바꾸고 행의 개수 세기
ns_book5.loc[old_books, '발행년도'] = -1
ns_book5['발행년도'].eq(-1).sum()  # -1과 같은 값이 몇 개인지 센다
```
