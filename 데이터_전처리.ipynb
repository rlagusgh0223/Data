위의 데이터를 변환해서 우리에게 필요한 데이터로 전환

```python
ad.info()
```

```python
ad['date'] = pd.to_datetime(ad.date)
ad.info()
```

```python
# 집계 단위를 week로 바꾸기
ad.date.dt.to_period("W")
```

```python
# 그 주의 시작하는 날짜로 변경
ad['week'] = ad.date.dt.to_period("W").dt.to_timestamp()
ad.tail()
```

```python
# 집계(단위는 weekly단위로 캠페인, spending이 몇번 있었는지 합산)
ad_agg = ad.groupby(by=['campaign', 'week'])[['spend', 'clicks']].sum().reset_index()
```

```python
ad_agg.head()
# CPC컬럼 만들기(Cost Per Click, 포스트를 클릭한 단위가 얼마인지)
ad_agg['cpc'] = ad_agg.spend/ad_agg.clicks
ad_agg.head()
```

Order에 log에서 유입경로 가져오기

```python
log.info()
```

```python
log['timestamp'] = pd.to_datetime(log.timestamp)
log.info()
```

```python
# log에서 campaign정보만 가져오기
log2 = log[~log.campaign.isna()]
```

```python
od.info()
```

```python
od['order_date'] = pd.to_datetime(od.order_date)
od.info()
```

```python
# 두 테이블에 order_id의 중복이 없는지
log.groupby('order_id').size().sort_values()
```

```python
od.groupby('order_id').size().sort_values()
```

```python
od1 = pd.merge(od, log, how='left', left_on='order_id', right_on='order_id')
od1.head()
```

```python
# campaign에 값이 없으면 organic으로 대체
od1['campaign'] = od1.campaign.fillna('organic')
od1.groupby(by='campaign').order_amount.sum()
```

```python
# orderdate기반으로 week만들기
od1['week'] = od1.order_date.dt.to_period('W').dt.to_timestamp()
```

```python
# 원하는 형태로 집계
od_agg = od1.groupby(by=['campaign', 'week'])['order_amount'].sum().reset_index()
```

광고지출과 매출데이터 병합하기

```python
ad_agg.head()
```

```python
od_agg.head()
```

```python
df = pd.merge(od_agg, ad_agg, how='left', left_on=['campaign', 'week'], right_on=['campaign', 'week'])
df.head()
```

```python
# 로아스 계산(클수록 비즈니스적으로 유리한 지표)
df['roas'] = df.order_amount/df.spend
df.head()
```

```python
# 위의 자료는 사람이 한번에 이해하기 어려우니
# 로아스가 위크별로 어떻게 변화되었나 피벗테이블 사용
df.pivot_table(
    index = 'campaign',
    columns = 'week',
    values = 'roas'
)
```

위의 계산을 계속하면 휴먼에러가 발생할 수 있다.
그래서 위의 과정들은 파이썬 스크립트는 한번 짜보고 자동화하기 좋다.
